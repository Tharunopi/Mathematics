```python
import matplotlib.pyplot as plt
```

```python
def vector_add(a, b):
    if len(a) != len(b):
        return print('invaild length!!')

    result = [a[i]+b[i] for i in range(len(a))]
    return print(result)


a = [1, 3]
b = [4, 7]
vector_add(a, b)
```

```python
def scaler_multi(s, v):
    result = [i*s for i in v]
    return print(result)

v = [2, 4]
s = 10
scaler_multi(s, v)
```

```python
def plot_vector(a):
    plt.quiver(0, 0, a[0], a[1], scale=1, angles='xy', scale_units='xy')

    plt.xlim([0, 10])
    plt.ylim([0, 10])

    plt.grid()
    plt.show()

plot_vector(a)
```
